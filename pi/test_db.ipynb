```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("chair.db")

# 查询所有有record_label的数据（新表）
df = pd.read_sql_query("""
    SELECT * FROM sensor_data 
    WHERE record_label IS NOT NULL
    ORDER BY timestamp DESC
""", conn)

print(f"Total records with record_label: {len(df)}")
print(df)

conn.close()
```
